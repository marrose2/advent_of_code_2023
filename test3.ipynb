```python
def find_number(i,j,visited):
    if ((j>=len(Lines[i])) or (j<=-1)):return "",visited
    if (not Lines[i][j].isnumeric()):return "",visited
    if (i==-1) or (i==len(Lines)): return "",visited
    k=j
    num_string=""

    while((k>-1) and (Lines[i][k].isnumeric())):
        
        num_string=Lines[i][k]+num_string
        if ((i,k) in visited): return "",visited
        visited.append((i,k))
        k-=1
    k=j+1
    while((k<len(Lines[i])) and (Lines[i][k].isnumeric())):
        num_string=num_string+Lines[i][k]
        if ((i,k) in visited): return "",visited
        visited.append((i,k))
        k+=1
    return num_string,visited
```
